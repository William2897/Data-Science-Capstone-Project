```python
import numpy as np 
import pandas as pd
import matplotlib.cm as cm
import matplotlib.colors as colors

import folium 
```

### Read saved Neighborhood data

```python
data = pd.read_csv('Location.csv')
```

```python
data.head()
```

### Drop Unneccessary columns

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
data.drop('Postal Code', axis=1, inplace=True)
```

data.head()

### Create map of Toronto using latitude and longitude values and add markers


```python
Toronto_latitude = 43.651070
Toronto_longitude = -79.347015
```

```python
Toronto_map = folium.Map(location=[Toronto_latitude, Toronto_longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(data['Latitude'], data['Longitude'], data['Borough'], data['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(Toronto_map)  
    
Toronto_map
```
